```python
import re

txt = '''Regular Expression is a method which is used in programming 
         to find pattern matching. For example we need to find out a 
         certain digit 123456789 from a text or (999)-333-7777 or any 
         other digit in that case we can use regular expression.'''

pattern = '\(\d{3}\)-\d{3}\-\d{4}'

   #here /d returns match of digit and \  used before ( ) to escape special character
   #as ( ) is also used to capture group so not to perform its function \ is provided before ()''
    
m = re.findall(pattern, txt)
m
```

```python
# lets find two given digit 123456789 or (999)-333-7777 together

txt = '''Regular Expression is a method which is used in programming 
         to find pattern matching. For example we need to find out a 
         certain digit 123456789 from a text or (999)-333-7777 or any 
         other digit in that case we can use regular expression.'''


pattern = "\d{9}|\(\d{3}\)-\d{3}\-\d{4}"
    
m = re.findall(pattern, txt)   # | symbol means or
print(m)
print(len(m))
```

```python
#lets look at some email id to extract

txt = '''Regular Expression is a method which is used in programming 
         to find pattern matching. For example we need to find out a 
         certain digit 123456789 from a text or (999)-333-7777 or any 
         other digit in that case we can use regular expression.
         Email: shibluar123@gmail.com and other email: abc@xyz.com and other email: AleX123@gmail.com 
         other email email: AleX_123@gmail.com'''


pattern = "[a-z 0-9A-Z_]*@[a-z0-9A-Z]*\.[a-z]*"
    
m = re.findall(pattern, txt)   # * indicates more occurance that means all character in a word 
                               #and we used \ before . as . also bears speacial meaning
print(m)
print(len(m))
```

**Let keep this information in a python dictionary**

```python
def get_match(pattern,text):
    
    matches = re.findall(pattern,text)
    
    if matches:
        return matches[0].strip()
    
##find name
get_match('Born(.*)',text)
```

```python
def get_information(text):
    
        Name = get_match('Born(.*)',text)
        Age =  get_match('age ([\d]*)',text)
        DOB =  get_match('Born.*\n(.*)\(age',text)
        Birth_Place = get_match('\(age.*\n(.*)',text)
        
        
        return{
            'Name':Name.strip(),
            'Age':int(Age),
            'Dob':DOB.strip(),
            'Birth_place': Birth_Place.strip()  
            
        }
        
#Lets call the function
get_information(text)
    
```
