**Stop Words** are the English words which does not add much meaning to a sentence. They can safely be ignored without sacrificing the meaning of the sentence. For example, the words like the, he, have etc. Such words are already captured this in corpus named corpus. We first download it to our python environment.

To know more stop words visit https://www.tutorialspoint.com/python_text_processing/python_remove_stopwords.htm

```python
import nltk, string

sampleText = '''My name is Alex. I study at Asia Pacific University.
                Currently I am in my final year and my final year project is about building a recomendation system engine. 
                After completing my bachelor degree in computer science, my planing is to pursue MBA. Is it a good decision?'''


#Lets first separate words from text using nltk and we made all words as lower case

text_lower = sampleText.lower()
wordtokens = nltk.tokenize.word_tokenize(text_lower)
print(wordtokens)

print("\nTotal number of words in this text corpus is ",len(wordtokens))
```

```python
#Lets print stopwords that are present in English language

stopTokens = nltk.corpus.stopwords.words("english") + list(string.punctuation)
print(stopTokens)
```

```python
#Lets see how many stop words are present in sample text and what are they

stopwordinText = []
for w in stopTokens:
    if w in wordtokens :
        stopwordinText.append(w)
        
print("\nThere are", len(stopwordinText), "stopwords in this text\n")
print(stopwordinText)   
```

**There are 15 stop words in the sample text now let remove this from sample text**

```python
filterToken = []

for word in wordtokens:
    if word not in stopTokens:
        filterToken.append(word)

print(filterToken)
```

**lets see length after and before removing stop words**

```python
print("Before removing stopwords from sampleText:  ",len(wordtokens))
print("\nAfter removing stopwords from sampleText:  ",len(filterToken))
```

```python
#Lets see 10 most common words and how many times these words appeared in sample text

fdist1 = nltk.FreqDist(filterToken)
print("\n",fdist1.most_common(10))
```

```python
#Lets see spacific words is present or not and how many times it appeared if it is present there

print("there are : ",wordtokens.count("skills"))
print("there are : ",wordtokens.count("final"))
```
