###### Tokenization

Tokenization is breaking the raw text into small chunks. Tokenization breaks the raw text into words, sentences called tokens. These tokens help in understanding the context or developing the model for the NLP. The tokenization helps in interpreting the meaning of the text by analyzing the sequence of the words.

Example:
when we put the sentence "This is a sample" into tokenization process we will have separate token like "This","is","a","sample".

**Lets have some example**

```python
#using regular expression we can split this text into separate words
#using whitespace we can separate the words

import re

text = '''Tokenization /is\ breaking the raw text into small chunks. 
          Tokenization breaks the raw text into words, sentences called tokens.
          These tokens help in understanding the context or developing the model for the NLP.'''


wordsbeforePlus = re.split(r'\W',text)
                # here r means that the string is to be treated as a raw string,
                #which means all escape codes will be ignored. Escape code like\,/,,\\,\t etc.
                #\W means it will return the string that does not contain any word characters
                #it will take special character like \,/," " etc. as empty string.
                        
print(wordsbeforePlus)

```

**lets run this again with adding + sign with \W**

```python
wordsAfterPlus = re.split(r'\W+',text)

# With adding a + wiht \W it will not take special character as string in words list

print(wordsAfterPlus)
 
#lets count the words in list before and after + sign 
countbeforeplus = 0;

print("There are ",len(wordsbeforePlus), "before adding plus sign") 
for i in wordsbeforePlus:
    countbeforeplus += 1
    print(countbeforeplus ,". ",wordsbeforePlus[countbeforeplus-1])


    
print("\nThere are" ,len(wordsAfterPlus), 'after adding plus sign\n')
countAfterplus = 0
for i in wordsAfterPlus:
    countAfterplus += 1
    print(countAfterplus ,". ",wordsAfterPlus[countAfterplus-1])
    
```

```python
import string
import re

#Split a string into a list where each word is a list item:
words = text.split()
print(words)

#function that re.compile does shown above. here it will return only the value we give
#The %s operator lets you add a value into a Python string and [] represent a set of character.
#re. escape(string) Return string with all non-alphanumerics backslashed; 
# So re.escape will not take string.punctuation sign.
#To know more about string.punctuation sign visit https://www.geeksforgeeks.org/string-punctuation-in-python/

re_punc = re.compile('[%s]' % re.escape(string.punctuation))

print(re_punc)

#Now remove punctuation form each words

#sub() function is used to replace occurrences of a particular sub-string with another sub-string.

#method 1
print("\nMethod 1")
method1 = [re_punc.sub('',w) for w in words]
print(method1)

#method 2
print("\nMethod 2")
method2 = []
for w in words:
    word = re_punc.sub('',w)
    method2.append(word)
    
print(method2)


#To understand sub method we can further do: 
print("\nsub function logic")
words = [re_punc.sub('/',w) for w in words]
print(words)
```

```python
#Let see re.escape function again to be more clear

Escape = re.compile('[%s]' % re.escape("/"))

print("Symbol attached in escape")
print(Escape)

#lets try sub function again.

#method 2
print("\nMethod 2")
method2 = []
for w in words:
    word = Escape.sub('',w)
    method2.append(word)
    
print(method2)
```

```python
#Lets do with string.printable inverse of string.punctution

text = "i\'m With you for the Entire life in U.S.A!"
words = text.split()
print(words)

re_punc = re.compile('[^%s]' % re.escape(string.printable))
print("\nMethod 1")
method1 = [re_punc.sub('',w) for w in words]
print(method1)
```

```python
#we can also make all words upper and lower case

words = text.split()

#Method 1
print("\nFirst Method")
wordList = []
for word in words:
    word = word.lower()
    wordList.append(word)
print(wordList)


#Mehtod 2
print("\n\nSecond Method") #New Line

word = [word.lower() for word in words]

print(word)
```

```python
#Lets import Spacy 

import spacy

#It will take a while to load
nlp = spacy.load('en_core_web_sm')
```

```python
#Lets write some text

text = '"i\'m study At Asia Pacific University which also know as A.P.U in short!"'

#Lets print the text first
print(text)


#Lets separate each words using loaded library

res = nlp(text)
print("\n")
for token in res:
     print(token.text,end ='|') #we have separated each word by | and we can now take character as words
    

    
# we can also take it in list separately

tokens = [token.text for token in res]
print("\n\nUsing List")
print(tokens)
```
